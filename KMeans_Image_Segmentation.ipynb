# KMeans Clustering

K-Means image segmentation is an algorithm that partitions pixels into clusters based on their feature similarity, typically color intensity values in images. The process involves iteratively assigning pixels to clusters represented by centroids, aiming to minimize the sum of squared distances between pixels and their respective cluster centroids. This iterative assignment continues until convergence criteria are met. The result is segmented regions where pixels within the same cluster exhibit similar color properties.

Load sample image

```python
img = mpl.image.imread(img_0)
```

View selected image

```python
plt.imshow(img)
```

the shape refers to the dimension of an image array representing the pixels height and width. where the height represent the number of rows in the image and width represent the number of column in the image.

```python
img.shape
```

reshape the image to a 2D array of pixels.

```python
X = img.reshape(-1, 3)
X.shape
```

perform kmeans clustering in pixel array.

```python
# modify the number of clusters
clusters = 4
```

initialize and fit the model.

```python
model = KMeans(n_clusters=clusters, n_init=10).fit(X)
```

replace pixel values with corresponding centroid values.

```python
segmented_img = model.cluster_centers_[model.labels_]
```

- reshape the flat array back to the original image shape.
